```python
import requests
from bs4 import BeautifulSoup
```

```python
url_to_scrape = "https://www.hamropatro.com/rashifal"
response = requests.get(url=url_to_scrape)
print(response.status_code)
```

```python
soup = BeautifulSoup(response.content,"html5lib")
print(soup.prettify())
```

```python
rashifals = soup.find_all("div",attrs={"class": "item"})
```

```python
rashifals
```

```python
for rashifal in rashifals:
    print(f"Rashi: {rashifal.h2.text}") # or rashifal.find("h3").text
    print(f"RashiImage: {rashifal.img['src']}") # or rashifal.img.get('src') or rashifal.find('img').get('src')
    print(f"Rashifal: {rashifal.p.text}")
```

```python
rashifal_dict = {}
for rashifal in rashifals:
    rashifal_dict[rashifal.h3.text] = {"image" : rashifal.img['src'],
                                        "desc": rashifal.p.text}
```
